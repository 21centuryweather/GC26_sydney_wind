```python
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
#Path g/data/gb02/mjl561
ds_u_win_DJF_1km_CTRL = xr.open_dataset('/g/data/gb02/mjl561/um2nc/SY_djf/SY_1/CTRL/wnd_ucmp-SY_SY_1_CTRL-v1-201611300100-201703010000.nc')
ds_u_win_DJF_1km_CTRL 
```

```python
u_win_DJF_1km_CTRL = ds_u_win_DJF_1km_CTRL['wnd_ucmp']
u_win_DJF_1km_CTRL
```

```python
#import level_height from ds_u_win_DJF_1km_CTRL
level_height = ds_u_win_DJF_1km_CTRL['level_height']
level_height
```

```python
print(level_height.values)
```

```python
#Path g/data/gb02/mjl561
ds_v_win_DJF_1km_CTRL = xr.open_dataset('/g/data/gb02/mjl561/um2nc/SY_djf/SY_1/CTRL/wnd_vcmp-SY_SY_1_CTRL-v1-201611300100-201703010000.nc')
ds_v_win_DJF_1km_CTRL 
```

```python
v_win_DJF_1km_CTRL = ds_v_win_DJF_1km_CTRL['wnd_vcmp']
v_win_DJF_1km_CTRL
```

```python
level_height2 = ds_v_win_DJF_1km_CTRL['level_height']
level_height2
```

```python
print(level_height2.values)
```

```python
#Calculating wind speed 
u_wind = u_win_DJF_1km_CTRL
v_wind = v_win_DJF_1km_CTRL
wind_speed = np.sqrt(u_wind**2+v_wind**2)
print (wind_speed)
```

```python
# --- Calculate time mean ---
wind_speed_mean = wind_speed.mean(dim='time')

# --- Plot ---
fig, ax = plt.subplots(
    figsize=(10, 8),
    subplot_kw={'projection': ccrs.PlateCarree()}
)

# Plot wind speed
im = ax.pcolormesh(
    wind_speed_mean.longitude,
    wind_speed_mean.latitude,
    wind_speed_mean,
    cmap='YlOrRd',
    transform=ccrs.PlateCarree()
)

# Add map features
ax.add_feature(cfeature.COASTLINE, linewidth=0.8)
ax.add_feature(cfeature.BORDERS, linewidth=0.5)
ax.add_feature(cfeature.STATES, linewidth=0.5, linestyle='--')

# Colorbar
cbar = plt.colorbar(im, ax=ax, orientation='vertical', pad=0.02, shrink=0.8)
cbar.set_label('Wind Speed (m/s)', fontsize=12)

# Labels and title
ax.set_title('Mean 10m Wind Speed — DJF CTRL (1km)', fontsize=14, fontweight='bold')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.gridlines(draw_labels=True, linewidth=0.3, color='gray', alpha=0.5)

plt.tight_layout()
plt.show()
```
